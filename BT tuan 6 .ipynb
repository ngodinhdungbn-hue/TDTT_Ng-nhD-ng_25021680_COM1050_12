```python
n=list(map(int,input().split()))
#Tao dict ghi so lan xuat hien cua so
so_lan={}
for i in range(len(n)):
    if n[i] not in so_lan:
        so_lan[n[i]]=1
    else:
        so_lan[n[i]]+=1
# Xac dinh so lan xuat hien nhieu nhat        
maaxi=0
for value in so_lan.values():
    if maaxi<=value:
        maaxi=value
#Tim so nho nhat co so lan xuat hien max
maxx=[]
for key,vlue in so_lan.items():
    if vlue==maaxi:
        maxx.append(int(key))
so_min=maxx[0]
for k in maxx:
    if k<so_min:
        so_min=k
print(so_min)
```
