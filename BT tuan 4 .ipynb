```python
def reverse(m):
    số_mới=0
    while not m<=0:
        số_mới=số_mới*10+m%10
        m=(m-m%10)/10
    return int(số_mới)
def xác_định_số_Palindrome(k):
    k=str(k)
    k=list(k)
    if len(k)%2==1:
        for i in range(0,int((len(k)-1)/2)):
            if i==(len(k)-1)/2-1 and k[i]==k[len(k)-1-i]:
                return True
            elif k[i]!=k[len(k)-i-1]:
                break
    if len(k)%2==0:
        for i in range(0,int(len(k)/2)):
            if i==(len(k)/2)-1 and k[int(len(k)/2)]==k[int((len(k)/2)-1)]:
                return True
            elif k[i]!=k[len(k)-1-i]:
                break
n=int(input('Nhập số nguyên dương n:'))
số_lần_lặp=0
while xác_định_số_Palindrome(n)!=True:
    số_lần_lặp+=1
    n=n+reverse(n)
else:
    print(số_lần_lặp,n)
    
        


```

```python

def Độ_dài_dãy_Collatz(m):
    k=1
    while m!=1:
        if m%2==0:
            m=m/2
            k+=1    
        else:
            m=m*3+1
            k+=1
    else: 
        return k
n=int(input('Nhập:'))
dộ_dài_max=0
for i in range(1,n+1):
    if Độ_dài_dãy_Collatz(i)>=dộ_dài_max:
        dộ_dài_max=Độ_dài_dãy_Collatz(i)
for h in range(1,n+1):
    if Độ_dài_dãy_Collatz(h)==dộ_dài_max:
        print(h,Độ_dài_dãy_Collatz(h))
        break

    
```
