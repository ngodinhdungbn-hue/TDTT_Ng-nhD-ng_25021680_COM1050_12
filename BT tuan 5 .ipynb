W5A3

```python
from math import sqrt
def is_prime(n:int):
    if n==1:
        return False
    elif n==2:
        return True
    else:
        for i in range(2,int(sqrt(n))+1):
            if n%i==0:
                return False
        else:
            return True
n=int(input('Nhập'))
print(is_prime(n))
```

```python
def so_hoan_hao(n,sum=None):
  if sum==None:
    sum=0
  for i in range(1,n):
    if n%i==0:
      sum=sum+i
  return True if sum==n else False
n=int(input('Nhập'))
print(so_hoan_hao(n))
```

```python
def w5a5(k,n=None):
  if n==None:
    n=[]
  for index,value in enumerate(n):
    if value==k:
      return index+1
  else:
    return -1
n=list(map(int,input().split()))
m=int(input())
print(w5a5(m,n))
```

```python
def may_tinh(a,c,b):
  a,b=float(a),float(b)
  if c=='+':
    return f"{a+b:.2f}"
  elif c=='-':
    return f"{a-b:.2f}"
  elif c=='*':
    return f"{a*b:.2f}"
  elif c=='/':
    return f"{a/b:.2f}"
a,b,c=input().split()
print(may_tinh(a,b,c))
```

```python
def ma_hoa(a,b):
  a=list(a)
  b=list(b)
  k={}
  d={}
  for i in range(len(a)):
      if a[i] not in k:
        k[a[i]]=b[i]
      else:
        if k[a[i]]!=b[i]:
          return False
  for i in k:
    d[k[i]]=k[i]
  if len(k)!=len(d):
    return False
  else:
    return True
a,b=input().split()
print(ma_hoa(a,b))
```
