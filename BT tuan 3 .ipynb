```python
n=float(input('Nhập một số'))
def làm_tròn_số_dương(n):
    m=n-int(n)
    if m>=0.5:
        m=1
        n=n+1
    else:
        n=n
    return int(n)
def làm_tròn_số_âm(n):
    m=n-int(n)
    if m>-0.5:
        n=int(n)
    else:
        n=int(n)-1
    return int(n)
if n>=0:
    if n%int(n)==0:
        print(int(n),int(n),làm_tròn_số_dương(n))
    else:
        print(int(n)+1,int(n),làm_tròn_số_dương(n))
else:
    if n%int(n)==0:
        print(int(n),int(n),làm_tròn_số_âm(n))
    else:
        print(int(n),int(n)-1,làm_tròn_số_âm(n))
```
